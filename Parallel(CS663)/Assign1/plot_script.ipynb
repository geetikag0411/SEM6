```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
# Data to be plotted
data_label1 = [0.145, 0.188,  0.142, 0.156, 0.161, 0.130, 0.111, 0.116, 0.117]

data_label2 = [0.231,0.285,  0.155, 0.221, 0.272, 0.201, 0.211, 0.152, 0.180]
data_label3 = [1.92, 1.872,  1.414, 2.139,1.963, 1.739, 1.693, 1.947, 1.583]
data_label4 = [3.22, 2.837, 2.254, 3.472, 2.969,3.297, 2.918, 2.819, 3.072]

# Combine the data into a list of lists
data = [data_label1, data_label2, data_label3, data_label4]

# Create labels for x-axis
labels = ['512², 5', '512², 9', '2048², 5', '2048², 9']

# Create boxplots
fig, ax = plt.subplots(figsize=(10, 6))
plt.boxplot(data, labels=labels)

# Add title and labels
plt.title('Timing plots')
plt.xlabel('Data size (number of doubles), Stencil size')
plt.ylabel('Time (s)')
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(5))
ax.yaxis.grid(which="both", linewidth=0.25)
# Show plot
plt.grid(True)
plt.show()
```
